### Gráfico Cuantil-Cuantil (QQ Plot)

Sirve para ver gráficamente si un conjunto de datos se distribuye de cierta manera, a menos que se especifique de otra manera, el gráfico muestra si los datos se ajustan a una distribución normal. Los precios ordenados en el eje **y**, y en **x** los cuantiles teóricos y a cuántas desviaciones de la media están alejados estos valores.
La línea roja representa lo que deben seguir los puntos azules si se distribuyeran normalmente.

```python
import scipy.stats
```

```python
scipy.stats.probplot(data.nikkei, plot =plt)
plt.title("QQ Plot", size = 24 )
plt.show()
```

Ya hemos visto que los datos no se distribuyen normalmente.
