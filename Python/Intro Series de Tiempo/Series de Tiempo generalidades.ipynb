### Gr√°fico Cuantil-Cuantil (QQ Plot)

Sirve para ver gr√°ficamente si un conjunto de datos se distribuye de cierta manera, a menos que se especifique de otra manera, el gr√°fico muestra si los datos se ajustan a una distribuci√≥n normal. Los precios ordenados en el eje **y**, y en **x** los cuantiles te√≥ricos y a cu√°ntas desviaciones de la media est√°n alejados estos valores.
La l√≠nea roja representa lo que deben seguir los puntos azules si se distribuyeran normalmente.

```python
import scipy.stats
```

```python
scipy.stats.probplot(data.nikkei, plot =plt)
plt.title("QQ Plot", size = 24 )
plt.show()
```

Ya hemos visto que los datos no se distribuyen normalmente.
